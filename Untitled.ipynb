```python
import pandas as p
import numpy as n
import matplotlib.pyplot as m
import seaborn as s

da=p.read_csv("C:\\Users\\reddy\\Desktop\\Dataset\\Salary Experience.csv")

da.dropna(axis=0,inplace=True)

x=da['Years of Experience'].values.reshape(-1,1)
y=da['Salary']

m.figure(figsize=(18,10))
m.subplot(1,2,1)
s.distplot(da['Years of Experience'])
m.subplot(1,2,2)
s.distplot(da['Salary'])
m.show()

m.scatter(x,y)
m.show()

from sklearn.model_selection import train_test_split
xtr,xt,ytr,yt=train_test_split(x,y,test_size=0.3,random_state=42)

from sklearn.linear_model import LinearRegression
mod=LinearRegression()
mod.fit(xtr,ytr)
yp=mod.predict(xt)

m.scatter(xt,yt,color='cyan',alpha=0.5,label="Actual Data Points")
m.plot(xt,yp,color="k",linewidth=0.3, label="Regression Line")
m.legend()
m.show()

from sklearn.metrics import r2_score,mean_squared_error
print("R2 Score:",r2_score(yt,yp))
print("RSME:",n.sqrt(mean_squared_error(yt,yp)))
```

```python
import pandas as p
import matplotlib.pyplot as m
da=p.read_csv("C:\\Users\\reddy\\Desktop\\Data\\Mall_Customers.csv")
x=da.iloc[:,[3,4]].values
from sklearn.cluster import KMeans
li=[]

for i in range(1,11):
    mod=KMeans(n_clusters=i,init='k-means++', random_state=42,n_init=10)
    mod.fit(x)
    li.append(mod.inertia_)

m.plot(range(1,11), li)
m.title("Elbow Method")
m.xlabel('Number of Clusters'); m.ylabel('li'); m.show()

mod=KMeans(n_clusters=5,init='k-means++', random_state=42,n_init=10)
y_pred=mod.fit_predict(x)

colors = ['blue', 'g', 'lime', 'red', 'k']
for cluster_num in range(5):
    m.scatter(x[y_pred == cluster_num, 0], x[y_pred == cluster_num, 1], s=100, c=colors[cluster_num], label=f'Cluster {cluster_num+1}')

m.scatter (mod.cluster_centers_[:,0],mod.cluster_centers_[:,1],s=300,c='yellow')
m.title("clusters of customers")
m.xlabel("Annual Income"); m.ylabel("Spending Score(1-100)")
m.legend();m.show()
```

```python
import tensorflow as t
import matplotlib.pyplot as m
from keras.utils import to_categorical
import keras as k

l=k.layers

from keras.datasets import mnist
(xtr,ytr),(xt,yt)=mnist.load_data()

xtr=xtr.astype('float32')/255.0
xt=xt.astype('float32')/255.0

ytr=to_categorical(ytr,10)
yt=to_categorical(yt,10)

for i in range(1,16):
    m.subplot(5,5,i)
    m.imshow(xtr[i], cmap=m.get_cmap('viridis'))

mod=k.Sequential()
mod.add(l.Flatten())
mod.add(l.Dense(5,activation='sigmoid'))
mod.add(l.Dense(10,activation='softmax'))

mod.compile(optimizer='adam', loss='categorical_crossentropy',metrics='accuracy')
mod.fit(xtr,ytr,batch_size=32, epochs=10, validation_data=[xt,yt])

lo,acc=mod.evaluate(xt,yt)
print("Loss:",lo)
print("Accuracy:",acc)
```

```python
import matplotlib.pyplot as m
import tensorflow as t
from keras.utils import to_categorical

from keras.datasets import mnist
(xtr,ytr),(xt,yt)=mnist.load_data()

xtr=xtr.astype('float32')/255.0
xt=xt.astype('float32')/255.0

ytr=to_categorical(ytr)
yt=to_categorical(yt)

for i in range(1,16):
    m.subplot(5,5,i)
    m.imshow(xtr[i])
m.show()

l=t.keras.layers
mod=t.keras.Sequential()
mod.add(l.Flatten())
mod.add(l.Dense(5,activation='sigmoid'))
mod.add(l.Dense(10,activation='softmax'))

mod.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
mod.fit(xtr,ytr,epochs=10,batch_size=32,validation_data=[xt,yt])

lo,acc=mod.evaluate(xt,yt)
print("Loss:",lo)
print("Accuracy:",acc)
```

```python
import pandas as p
import seaborn as s
from sklearn.datasets import load_diabetes

dp=load_diabetes()
co=dp.feature_names
da=p.DataFrame(dp.data);da.columns=co

s.boxplot(data=da,palette='Set1')

q1=da['bmi'].quantile(0.25)
q3=da['bmi'].quantile(0.75)
iqr=q3-q1
flo=1.5-q1*iqr
cap=1.5+q3*iqr
out=(da.bmi<=flo) | (da.bmi>=cap)
```
